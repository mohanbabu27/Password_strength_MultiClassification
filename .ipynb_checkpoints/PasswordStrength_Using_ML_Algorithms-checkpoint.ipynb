```python
def word_devide_char(inputs):
    characters=[] #Whatever the inputs giving, its stores each character by character in list
    for i in inputs:
        characters.append(i)
        return characters
```

```python

from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer(tokenizer=word_devide_char)
X=vectorizer.fit_transform(X)
```

```python
feature_names = vectorizer.get_feature_names()
#get tfidf vector for first document
first_document_vector=X[0]
#print the scores
df = pd.DataFrame(first_document_vector.T.todense(), index=feature_names, columns=["tfidf"])
df.sort_values(by=["tfidf"],ascending=False)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression 
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
log_class=LogisticRegression(penalty='l2', multi_class='ovr')
log_class.fit(X_train,y_train)
```

```python
print(log_class.score(X_test,y_test))
```

```python
##Incase of Multinominal
clf = LogisticRegression(random_state=0, multi_class='multinomial', solver='newton-cg')
clf.fit(X_train, y_train)
print(clf.score(X_test, y_test))
```

```python
X_predict=np.array(["qwer34nsf567%^&"])
X_predict=vectorizer.transform(X_predict)
y_pred=log_class.predict(X_predict)
print(y_pred)
```
